## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python


# ===============================================================
# IMPORTS
# ===============================================================
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import random
from sklearn.metrics import r2_score, mean_absolute_error
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model

np.random.seed(0)
tf.random.set_seed(0)

# ===============================================================
# LOAD DATA FUNCTION
# ===============================================================
def extract_cycles(path):
    df = pd.read_csv(path)
    df.columns = df.columns.str.strip()
    curves, soh = [], []
    Qref = None
    for cyc in sorted(df['cycle number'].unique()):
        c = df[df['cycle number'] == cyc]
        dis = c[c['<I>/mA'] < 0]
        if len(dis) < 80: continue
        V = dis['Ecell/V'].values
        Q = dis['Q discharge/mA.h'].values
        if Q.max() <= 0: continue
        if Qref is None: Qref = Q.max()
        s = Q.max() / Qref
        if s < 0.7: continue
        Vn = (V - V.min()) / (V.max() - V.min())
        idx = np.argsort(Vn)
        Qi = np.interp(np.linspace(0,1,400), Vn[idx], Q[idx])
        curves.append(Qi / Qref)
        soh.append(s)
    return np.array(curves), np.array(soh)

# ===============================================================
# FILTER VALID FILES (≥300 cycles)
# ===============================================================
all_files = [os.path.join(DATA_PATH, f) for f in os.listdir(DATA_PATH) if f.endswith(".csv")]
valid_files = []
for f in all_files:
    try:
        df = pd.read_csv(f)
        if df['cycle number'].nunique() >= 300:
            valid_files.append(f)
    except Exception as e:
        print(f"Skipping {f} due to error: {e}")

print("Total valid files:", len(valid_files))

# Split: 25 train, 5 test
random.seed(0)
test_files = random.sample(valid_files, 5)
train_files = [f for f in valid_files if f not in test_files]

print("Training files:", len(train_files))
print("Testing files:", len(test_files))

# ===============================================================
# PREPARE TRAINING DATA
# ===============================================================
Q_train_all, soh_train_all, first_caps = [], [], []
for f in train_files:
    Q_tmp, soh_tmp = extract_cycles(f)
    if len(Q_tmp) == 0: continue
    Q_train_all.append(Q_tmp)
    soh_train_all.append(soh_tmp)
    first_caps.append(np.trapezoid(Q_tmp[0], np.linspace(0,1,400)))

Q_train = np.vstack(Q_train_all)
soh_train = np.concatenate(soh_train_all)

# ===============================================================
# LIPM MODEL + PARAM FITTING
# ===============================================================
V_common = np.linspace(0,1,400)
dV = V_common[1] - V_common[0]
V_tf = tf.constant(V_common.reshape(1,-1), dtype=tf.float32)

def lipm(V,A1,V01,w1,A2,V02,w2,C,k):
    w1=np.abs(w1)+1e-6; w2=np.abs(w2)+1e-6; k=-np.abs(k)
    return (A1/np.pi)*np.arctan((2*(V-V01))/w1) + \
           (A2/np.pi)*np.arctan((2*(V-V02))/w2) + \
           C*np.exp(k*V)

def lipm_tf(V,p):
    A1=p[:,0:1]; V01=p[:,1:2]; w1=tf.abs(p[:,2:3])+1e-6
    A2=p[:,3:4]; V02=p[:,4:5]; w2=tf.abs(p[:,5:6])+1e-6
    C=p[:,6:7]; k=-tf.abs(p[:,7:8])
    Q=(A1/np.pi)*tf.atan((2*(V-V01))/w1)
    Q+=(A2/np.pi)*tf.atan((2*(V-V02))/w2)
    Q+=C*tf.exp(k*V)
    return Q

def fit_params(Q):
    plist=[]
    for q in Q:
        try:
            p,_=curve_fit(lipm, V_common, q,
                          p0=[1,.3,.1,1,.7,.1,.1,1],
                          maxfev=20000)
            plist.append(p)
        except: continue
    return np.array(plist)

params_train = fit_params(Q_train)

# ===============================================================
# FUTURE HORIZON + NORMALIZATION
# ===============================================================
future_horizon = 300
X, Y = [], []
for i in range(len(params_train) - future_horizon):
    X.append(params_train[i])
    Y.append(params_train[i+1:i+future_horizon+1].reshape(-1))
X = np.array(X, dtype=np.float32)
Y = np.array(Y, dtype=np.float32)

Xmean, Xstd = X.mean(0), X.std(0) + 1e-8
Ymean, Ystd = Y.mean(0), Y.std(0) + 1e-8
Xn = (X - Xmean)/Xstd
Yn = (Y - Ymean)/Ystd

# ===============================================================
# NETWORK
# ===============================================================
inp = Input(shape=(8,))
x = Dense(128, activation='tanh', kernel_regularizer=tf.keras.regularizers.l2(1e-6))(inp)
x = Dropout(0.1)(x)
x = Dense(256, activation='tanh')(x)
x = Dense(128, activation='tanh', kernel_regularizer=tf.keras.regularizers.l2(1e-6))(x)
out = Dense(future_horizon*8)(x)
model = Model(inp, out)
optimizer = tf.keras.optimizers.Adam(1e-4)

# ===============================================================
# GLOBAL BASELINE CAPACITY
# ===============================================================
baseline_capacity = np.mean(first_caps)
baseline_capacity_tf = tf.constant(baseline_capacity, dtype=tf.float32)

# ===============================================================
# HYBRID LOSS
# ===============================================================
@tf.function
def train_step(x_batch,y_batch):
    with tf.GradientTape() as tape:
        y_pred = model(x_batch, training=True)
        data_loss = tf.reduce_mean(tf.square(y_batch - y_pred))

        pp = y_pred * Ystd + Ymean
        tp = y_batch * Ystd + Ymean
        pp = tf.reshape(pp, (-1, future_horizon, 8))
        tp = tf.reshape(tp, (-1, future_horizon, 8))
        pp_flat = tf.reshape(pp, (-1,8))
        tp_flat = tf.reshape(tp, (-1,8))

        Qp = lipm_tf(V_tf, pp_flat)
        Qt = lipm_tf(V_tf, tp_flat)
        phy_loss = tf.reduce_mean(tf.square(Qp - Qt))

        cap_pred = tf.reduce_sum((Qp[:,1:] + Qp[:,:-1]) * 0.5 * dV, axis=1)
        cap_true = tf.reduce_sum((Qt[:,1:] + Qt[:,:-1]) * 0.5 * dV, axis=1)

        soh_pred = cap_pred / baseline_capacity_tf
        soh_true = cap_true / baseline_capacity_tf
        soh_loss = tf.reduce_mean(tf.square(soh_true - soh_pred))

        cap_pred_seq = tf.reshape(cap_pred, (-1, future_horizon))
        upper_penalty = tf.reduce_mean(tf.nn.relu(soh_pred - 1.0))
        mono_penalty = tf.reduce_mean(tf.nn.relu(cap_pred_seq[:,1:] - cap_pred_seq[:,:-1]))
        constraint_loss = 0.5*upper_penalty + 0.7*mono_penalty

        loss = 0.4*data_loss + 0.4*phy_loss + 0.6*soh_loss + constraint_loss

    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss

# ===============================================================
# TRAINING LOOP
# ===============================================================
dataset = tf.data.Dataset.from_tensor_slices((Xn,Yn)).shuffle(len(Xn)).batch(64)

for epoch in range(500):
    total=0; steps=0
    for xb,yb in dataset:
        l=train_step(xb,yb)
        total+=l.numpy(); steps+=1
    if epoch % 20 == 0:
        print("Epoch", epoch, "Loss:", total/steps)

# ===============================================================
# EVALUATION FUNCTION
# ===============================================================
def evaluate_on_file(file_path, cycle_index=10):
    Q_new, soh_new = extract_cycles(file_path)
    params_new = fit_params(Q_new)

    if len(params_new) <= cycle_index:
        print(f"Skipping {file_path} (not enough cycles)")
        return None

    test_input = ((params_new[cycle_index] - Xmean)/Xstd).reshape(1,-1).astype(np.float32)
    pred = model.predict(test_input)[0]
    pred = pred*Ystd + Ymean
    pred = pred.reshape(future_horizon,8)

    pred_soh = []
    for p in pred:
        curve = lipm(V_common,*p)
        curve = np.clip(curve,0,None)
        capacity = np.trapezoid(curve,V_common)
        pred_soh.append(capacity/baseline_capacity)
    pred_soh = np.clip(np.array(pred_soh), 0, 1.0)

    true_soh_future = soh_new[cycle_index+1:cycle_index+1+future_horizon]
    L = min(len(true_soh_future), len(pred_soh))
    if L == 0:
        print(f"Skipping {file_path} (no future cycles)")
        return None

    r2 = r2_score(true_soh_future[:L], pred_soh[:L])
    mae = mean_absolute_error(true_soh_future[:L], pred_soh[:L])
    rmse = np.sqrt(np.mean((true_soh_future[:L] - pred_soh[:L])**2))

    print(f"Results for {os.path.basename(file_path)}")
    print("R2:", r2, "MAE:", mae, "RMSE:", rmse)
    print("-"*50)

    # Plot trajectory
    plt.figure(figsize=(10,6))
    plt.plot(range(cycle_index+1, cycle_index+1+L),
             true_soh_future[:L], label="True SOH")
    plt.plot(range(cycle_index+1, cycle_index+1+L),
             pred_soh[:L], '--', label="Predicted SOH")
    plt.xlabel("Cycle")
    plt.ylabel("SOH")
    plt.title(f"{os.path.basename(file_path)} Prediction (R2={r2:.3f})")
    plt.grid()
    plt.legend()
    plt.show()

    return r2, mae, rmse

# ===============================================================
# LOOP THROUGH ALL TEST FILES
# ===============================================================
results = {}
for f in test_files:
    res = evaluate_on_file(f, cycle_index=10)
    if res is not None:
        results[f] = res

print("Summary of results:")
for f, (r2, mae, rmse) in results.items():
    print(f"{os.path.basename(f)} -> R2: {r2:.3f}, MAE: {mae:.4f}, RMSE: {rmse:.4f}")

````

````python
#may not work--trying other way ===============================================================
# IMPORTS
# ===============================================================
from scipy.optimize import curve_fit
import numpy as np, pandas as pd, tensorflow as tf, matplotlib.pyplot as plt, os
from sklearn.metrics import r2_score, mean_absolute_error
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model

np.random.seed(0); tf.random.set_seed(0)

# ===============================================================
# DATA PATH
# ===============================================================
DATA_PATH = "/content/Dataset_1_NCA_battery/Dataset_1_NCA_battery"
print("CSV files:", len([f for f in os.listdir(DATA_PATH) if f.endswith(".csv")]))

# ===============================================================
# LOAD DATA FUNCTION
# ===============================================================
def extract_cycles(path):
    df = pd.read_csv(path); df.columns = df.columns.str.strip()
    curves, soh = [], []; Qref = None
    for cyc in sorted(df['cycle number'].unique()):
        c = df[df['cycle number'] == cyc]; dis = c[c['<I>/mA'] < 0]
        if len(dis) < 80: continue
        V = dis['Ecell/V'].values; Q = dis['Q discharge/mA.h'].values
        if Q.max() <= 0: continue
        if Qref is None: Qref = Q.max()
        s = Q.max()/Qref
        if s < 0.7: continue
        Vn = (V - V.min())/(V.max()-V.min()); idx = np.argsort(Vn)
        Qi = np.interp(np.linspace(0,1,400), Vn[idx], Q[idx])
        curves.append(Qi/Qref); soh.append(s)
    return np.array(curves), np.array(soh)

# ===============================================================
# FILTER VALID FILES
# ===============================================================
all_files = [os.path.join(DATA_PATH,f) for f in os.listdir(DATA_PATH) if f.endswith(".csv")]
valid_files = []
for f in all_files:
    try:
        df = pd.read_csv(f)
        if df['cycle number'].nunique() >= 300: valid_files.append(f)
    except: continue
print("Total valid files:", len(valid_files))

# ===============================================================
# GROUP FILES BY CONDITION (temperature, C-rate)
# ===============================================================
def parse_condition(fname):
    fname = fname.lower()
    # Temperature
    if "CY25" in fname: temp = 25.0
    elif "CY35" in fname: temp = 35.0
    elif "CY45" in fname: temp = 45.0
    else: temp = 25.0  # default

    # C-rate
    if "-1_" in fname: c_rate = 1.0
    elif "-05_" in fname: c_rate = 0.5   # handles "05"
    elif "-025_" in fname: c_rate = 0.25 # handles "025"
    else: c_rate = 1.0

    return temp, c_rate


condition_groups = {}
for f in valid_files:
    temp, c_rate = parse_condition(os.path.basename(f))
    condition_groups.setdefault((temp,c_rate), []).append(f)

print("Condition groups:", {k:len(v) for k,v in condition_groups.items()})

# Example: train on 25°C & 35°C, test on 45°C
train_conditions = [(25.0,1.0),(25.0,0.5),(25.0,0.25),(35.0,1.0),(35.0,0.5),(35.0,0.25)]
test_conditions  = [(45.0,1.0),(45.0,0.5),(45.0,0.25)]

train_files = sum([condition_groups[c] for c in train_conditions if c in condition_groups], [])
test_files  = sum([condition_groups[c] for c in test_conditions if c in condition_groups], [])

print("Training files:", len(train_files), "Testing files:", len(test_files))

# ===============================================================
# LIPM MODEL
# ===============================================================
V_common = np.linspace(0,1,400); dV = V_common[1]-V_common[0]
V_tf = tf.constant(V_common.reshape(1,-1),dtype=tf.float32)

def lipm(V,A1,V01,w1,A2,V02,w2,C,k):
    w1=np.abs(w1)+1e-6; w2=np.abs(w2)+1e-6; k=-np.abs(k)
    return (A1/np.pi)*np.arctan((2*(V-V01))/w1) + \
           (A2/np.pi)*np.arctan((2*(V-V02))/w2) + \
           C*np.exp(k*V)

def lipm_tf(V,p):
    A1=p[:,0:1]; V01=p[:,1:2]; w1=tf.abs(p[:,2:3])+1e-6
    A2=p[:,3:4]; V02=p[:,4:5]; w2=tf.abs(p[:,5:6])+1e-6
    C=p[:,6:7]; k=-tf.abs(p[:,7:8])
    Q=(A1/np.pi)*tf.atan((2*(V-V01))/w1)
    Q+=(A2/np.pi)*tf.atan((2*(V-V02))/w2)
    Q+=C*tf.exp(k*V)
    return Q

def fit_params(Q):
    plist=[]
    for q in Q:
        try:
            p,_=curve_fit(lipm,V_common,q,p0=[1,.3,.1,1,.7,.1,.1,1],maxfev=20000)
            plist.append(p)
        except: continue
    return np.array(plist)

# ===============================================================
# CREATE TRAINING DATA (with condition inputs)
# ===============================================================
future_horizon = 300; X=[]; Y=[]; first_caps=[]
for f in train_files:
    Q_tmp, soh_tmp = extract_cycles(f); params_tmp = fit_params(Q_tmp)
    if len(params_tmp) <= future_horizon: continue
    temp,c_rate = parse_condition(os.path.basename(f))
    for i in range(len(params_tmp)-future_horizon):
        X.append(np.concatenate([params_tmp[i],[temp,c_rate]]))
        Y.append(params_tmp[i+1:i+future_horizon+1].reshape(-1))
    first_caps.append(np.trapezoid(Q_tmp[0], np.linspace(0,1,400)))

X=np.array(X,dtype=np.float32); Y=np.array(Y,dtype=np.float32)
print("Training samples:",X.shape)

# ===============================================================
# NORMALIZATION
# ===============================================================
Xmean,Xstd = X.mean(0),X.std(0)+1e-8
Ymean,Ystd = Y.mean(0),Y.std(0)+1e-8
Xn=(X-Xmean)/Xstd; Yn=(Y-Ymean)/Ystd

# ===============================================================
# NETWORK
# ===============================================================
inp = Input(shape=(10,))
x = Dense(128,activation='tanh',kernel_regularizer=tf.keras.regularizers.l2(1e-6))(inp)
x = Dropout(0.1)(x)
x = Dense(256,activation='tanh')(x)
x = Dense(128,activation='tanh',kernel_regularizer=tf.keras.regularizers.l2(1e-6))(x)
out = Dense(future_horizon*8)(x)
model = Model(inp,out)
optimizer = tf.keras.optimizers.Adam(1e-4)

# ===============================================================
# BASELINE CAPACITY
# ===============================================================
baseline_capacity = np.mean(first_caps)
baseline_capacity_tf = tf.constant(baseline_capacity,dtype=tf.float32)

# ===============================================================
# TRAIN STEP
# ===============================================================
@tf.function
def train_step(x_batch,y_batch):
    with tf.GradientTape() as tape:
        y_pred = model(x_batch,training=True)
        data_loss = tf.reduce_mean(tf.square(y_batch-y_pred))
        pp = y_pred*Ystd + Ymean; tp = y_batch*Ystd + Ymean
        pp = tf.reshape(pp,(-1,future_horizon,8)); tp = tf.reshape(tp,(-1,future_horizon,8))
        pp_flat = tf.reshape(pp,(-1,8)); tp_flat = tf.reshape(tp,(-1,8))
        Qp = lipm_tf(V_tf,pp_flat); Qt = lipm_tf(V_tf,tp_flat)
        phy_loss = tf.reduce_mean(tf.square(Qp-Qt))
        cap_pred = tf.reduce_sum((Qp[:,1:]+Qp[:,:-1])*0.5*dV,axis=1)
        cap_true = tf.reduce_sum((Qt[:,1:]+Qt[:,:-1])*0.5*dV,axis=1)
        soh_loss = tf.reduce_mean(tf.square(soh_true-soh_pred))
        cap_pred_seq = tf.reshape(cap_pred,(-1,future_horizon))
        upper_penalty = tf.reduce_mean(tf.nn.relu(soh_pred-1.0))
        mono_penalty = tf.reduce_mean(tf.nn.relu(cap_pred_seq[:,1:] - cap_pred_seq[:,:-1]))
        constraint_loss = 1.0*upper_penalty + 0.7*mono_penalty
        loss = 0.4*data_loss + 0.4*phy_loss + 0.6*soh_loss + constraint_loss

    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss

# ===============================================================
# TRAINING LOOP
# ===============================================================
dataset = tf.data.Dataset.from_tensor_slices((Xn,Yn)).shuffle(len(Xn)).batch(64)

for epoch in range(500):
    total=0; steps=0
    for xb,yb in dataset:
        l=train_step(xb,yb)
        total+=l.numpy(); steps+=1
    if epoch % 50 == 0:
        print("Epoch", epoch, "Loss:", total/steps)

# ===============================================================
# EVALUATION FUNCTION
# ===============================================================
def evaluate_on_file(file_path, cycle_index=10):
    Q_new, soh_new = extract_cycles(file_path)
    params_new = fit_params(Q_new)
    if len(params_new) <= cycle_index:
        print(f"Skipping {file_path} (not enough cycles)")
        return None

    temp,c_rate = parse_condition(os.path.basename(file_path))
    test_input = np.concatenate([params_new[cycle_index],[temp,c_rate]])
    test_input = ((test_input - Xmean)/Xstd).reshape(1,-1).astype(np.float32)

    pred = model.predict(test_input)[0]
    pred = pred*Ystd + Ymean
    pred = pred.reshape(future_horizon,8)

    pred_soh=[]
    for p in pred:
        curve = lipm(V_common,*p)
        curve = np.clip(curve,0,None)
        capacity = np.trapezoid(curve,V_common)
        pred_soh.append(capacity/baseline_capacity)
    pred_soh = np.clip(np.array(pred_soh),0,1)

    true_soh_future = soh_new[cycle_index+1:cycle_index+1+future_horizon]
    L = min(len(true_soh_future),len(pred_soh))
    if L == 0: return None

    r2 = r2_score(true_soh_future[:L],pred_soh[:L])
    mae = mean_absolute_error(true_soh_future[:L],pred_soh[:L])
    rmse = np.sqrt(np.mean((true_soh_future[:L]-pred_soh[:L])**2))

    print(os.path.basename(file_path),"-> R2:",r2," MAE:",mae," RMSE:",rmse)
    return r2,mae,rmse

# ===============================================================
# RUN TESTS (condition-based)
# ===============================================================
results={}
for f in test_files:
    res = evaluate_on_file(f)
    if res is not None: results[f]=res

print("\nSummary")
for f,(r2,mae,rmse) in results.items():
    print(os.path.basename(f),"-> R2:",round(r2,3)," MAE:",round(mae,4)," RMSE:",round(rmse,4))
````

````python
######using one model with features as inputs
# ===============================================================

# DOWNLOAD DATASET

# ===============================================================

!wget -O Dataset_1_NCA_battery.zip "https://zenodo.org/records/6405084/files/Dataset_1_NCA_battery.zip?download=1"

!unzip -q Dataset_1_NCA_battery.zip -d /content/Dataset_1_NCA_battery


# ===============================================================

# IMPORTS

# ===============================================================

from scipy.optimize import curve_fit

import numpy as np

import pandas as pd

import tensorflow as tf

import matplotlib.pyplot as plt

import os, random

from sklearn.metrics import r2_score, mean_absolute_error

from tensorflow.keras.layers import Input, Dense, Dropout

from tensorflow.keras.models import Model

np.random.seed(0)

tf.random.set_seed(0)


# ===============================================================

# DATA PATH

# ===============================================================

DATA_PATH = "/content/Dataset_1_NCA_battery/Dataset_1_NCA_battery"

print("CSV files:", len([f for f in os.listdir(DATA_PATH) if f.endswith(".csv")]))


# ===============================================================

# LOAD DATA FUNCTION

# ===============================================================

def extract_cycles(path):

    df = pd.read_csv(path)

    df.columns = df.columns.str.strip()

    curves, soh = [], []

    Qref = None

    for cyc in sorted(df['cycle number'].unique()):

        c = df[df['cycle number'] == cyc]

        dis = c[c['<I>/mA'] < 0]

        if len(dis) < 80:

            continue

        V = dis['Ecell/V'].values

        Q = dis['Q discharge/mA.h'].values

        if Q.max() <= 0:

            continue

        if Qref is None:

            Qref = Q.max()

        s = Q.max() / Qref

        if s < 0.7:

            continue

        Vn = (V - V.min()) / (V.max() - V.min())

        idx = np.argsort(Vn)

        Qi = np.interp(np.linspace(0,1,400), Vn[idx], Q[idx])

        curves.append(Qi / Qref)

        soh.append(s)

    return np.array(curves), np.array(soh)


# ===============================================================

# PARSE TEMPERATURE AND C-RATE FROM FILE NAME

# ===============================================================

def parse_condition(fname):

    fname = fname.lower()

    if "cy25" in fname:

        temp = 25.0

    elif "cy35" in fname:

        temp = 35.0

    elif "cy45" in fname:

        temp = 45.0

    else:

        temp = 25.0

    if "-1_" in fname:

        c_rate = 1.0

    elif "-05_" in fname:

        c_rate = 0.5

    elif "-025_" in fname:

        c_rate = 0.25

    else:

        c_rate = 1.0

    return temp, c_rate


# ===============================================================

# FILTER VALID FILES

# ===============================================================

all_files = [os.path.join(DATA_PATH,f) for f in os.listdir(DATA_PATH) if f.endswith(".csv")]

valid_files = []

for f in all_files:

    try:

        df = pd.read_csv(f)

        if df['cycle number'].nunique() >= 300:

            valid_files.append(f)

    except:

        continue

print("Total valid files:",len(valid_files))


# ===============================================================

# TRAIN TEST SPLIT

# ===============================================================

random.seed(0)

test_files = random.sample(valid_files,5)

train_files = [f for f in valid_files if f not in test_files]

print("Training files:",len(train_files))

print("Testing files:",len(test_files))


# ===============================================================

# LIPM MODEL

# ===============================================================

V_common = np.linspace(0,1,400)

dV = V_common[1] - V_common[0]

V_tf = tf.constant(V_common.reshape(1,-1),dtype=tf.float32)


def lipm(V,A1,V01,w1,A2,V02,w2,C,k):

    w1=np.abs(w1)+1e-6

    w2=np.abs(w2)+1e-6

    k=-np.abs(k)

    return (A1/np.pi)*np.arctan((2*(V-V01))/w1) + \
           (A2/np.pi)*np.arctan((2*(V-V02))/w2) + \
           C*np.exp(k*V)


def lipm_tf(V,p):

    A1=p[:,0:1]; V01=p[:,1:2]; w1=tf.abs(p[:,2:3])+1e-6

    A2=p[:,3:4]; V02=p[:,4:5]; w2=tf.abs(p[:,5:6])+1e-6

    C=p[:,6:7]; k=-tf.abs(p[:,7:8])

    Q=(A1/np.pi)*tf.atan((2*(V-V01))/w1)

    Q+=(A2/np.pi)*tf.atan((2*(V-V02))/w2)

    Q+=C*tf.exp(k*V)

    return Q


def fit_params(Q):

    plist=[]

    for q in Q:

        try:

            p,_=curve_fit(

                lipm,

                V_common,

                q,

                p0=[1,.3,.1,1,.7,.1,.1,1],

                maxfev=20000

            )

            plist.append(p)

        except:

            continue

    return np.array(plist)


# ===============================================================

# CREATE TRAINING DATA

# ===============================================================

future_horizon = 300

X=[]

Y=[]

for f in train_files:

    Q_tmp, soh_tmp = extract_cycles(f)

    params_tmp = fit_params(Q_tmp)

    if len(params_tmp) <= future_horizon:

        continue

    temp,c_rate = parse_condition(os.path.basename(f))

    for i in range(len(params_tmp)-future_horizon):

        X.append(np.concatenate([params_tmp[i],[temp,c_rate]]))

        Y.append(params_tmp[i+1:i+future_horizon+1].reshape(-1))


X=np.array(X,dtype=np.float32)

Y=np.array(Y,dtype=np.float32)

print("Training samples:",X.shape)


# ===============================================================

# NORMALIZATION

# ===============================================================

Xmean,Xstd = X.mean(0),X.std(0)+1e-8

Ymean,Ystd = Y.mean(0),Y.std(0)+1e-8

Xn=(X-Xmean)/Xstd

Yn=(Y-Ymean)/Ystd


# ===============================================================

# NETWORK

# ===============================================================

inp = Input(shape=(10,))

x = Dense(128,activation='tanh',kernel_regularizer=tf.keras.regularizers.l2(1e-6))(inp)

x = Dropout(0.1)(x)

x = Dense(256,activation='tanh')(x)

x = Dense(128,activation='tanh',kernel_regularizer=tf.keras.regularizers.l2(1e-6))(x)

out = Dense(future_horizon*8)(x)

model = Model(inp,out)

optimizer = tf.keras.optimizers.Adam(1e-4)


# ===============================================================

# BASELINE CAPACITY

# ===============================================================

baseline_curve = lipm(V_common,*X[0][:8])

baseline_capacity = np.trapezoid(baseline_curve,V_common)

baseline_capacity_tf = tf.constant(baseline_capacity,dtype=tf.float32)


# ===============================================================

# TRAIN STEP

# ===============================================================

@tf.function

def train_step(x_batch,y_batch):

    with tf.GradientTape() as tape:

        y_pred = model(x_batch,training=True)

        data_loss = tf.reduce_mean(tf.square(y_batch-y_pred))

        pp = y_pred*Ystd + Ymean

        tp = y_batch*Ystd + Ymean

        pp = tf.reshape(pp,(-1,future_horizon,8))

        tp = tf.reshape(tp,(-1,future_horizon,8))

        pp_flat = tf.reshape(pp,(-1,8))

        tp_flat = tf.reshape(tp,(-1,8))

        Qp = lipm_tf(V_tf,pp_flat)

        Qt = lipm_tf(V_tf,tp_flat)

        phy_loss = tf.reduce_mean(tf.square(Qp-Qt))

        cap_pred = tf.reduce_sum((Qp[:,1:]+Qp[:,:-1])*0.5*dV,axis=1)

        cap_true = tf.reduce_sum((Qt[:,1:]+Qt[:,:-1])*0.5*dV,axis=1)

        soh_pred = cap_pred/baseline_capacity_tf

        soh_true = cap_true/baseline_capacity_tf

        soh_loss = tf.reduce_mean(tf.square(soh_true-soh_pred))

        cap_pred_seq = tf.reshape(cap_pred,(-1,future_horizon))

        upper_penalty = tf.reduce_mean(tf.nn.relu(soh_pred-1.0))

        mono_penalty = tf.reduce_mean(tf.nn.relu(cap_pred_seq[:,1:] - cap_pred_seq[:,:-1]))

        constraint_loss = 0.5*upper_penalty + 0.3*mono_penalty

        loss = 0.4*data_loss + 0.4*phy_loss + 0.6*soh_loss + constraint_loss

    grads = tape.gradient(loss,model.trainable_variables)

    optimizer.apply_gradients(zip(grads,model.trainable_variables))

    return loss


# ===============================================================

# TRAINING

# ===============================================================

dataset = tf.data.Dataset.from_tensor_slices((Xn,Yn)).shuffle(len(Xn)).batch(64)

for epoch in range(500):

    total=0

    steps=0

    for xb,yb in dataset:

        l=train_step(xb,yb)

        total+=l.numpy()

        steps+=1

    if epoch%50==0:

        print("Epoch",epoch,"Loss:",total/steps)


# ===============================================================

# EVALUATION

# ===============================================================

def evaluate_on_file(file_path,cycle_index=10):

    Q_new,soh_new = extract_cycles(file_path)

    params_new = fit_params(Q_new)

    temp,c_rate = parse_condition(os.path.basename(file_path))

    test_input = np.concatenate([params_new[cycle_index],[temp,c_rate]])

    test_input=((test_input-Xmean)/Xstd).reshape(1,-1).astype(np.float32)

    pred=model.predict(test_input)[0]

    pred=pred*Ystd+Ymean

    pred=pred.reshape(future_horizon,8)

    pred_soh=[]

    for p in pred:

        curve=lipm(V_common,*p)

        curve=np.clip(curve,0,None)

        capacity=np.trapezoid(curve,V_common)

        pred_soh.append(capacity/baseline_capacity)

    pred_soh=np.array(pred_soh)

    true_soh_future=soh_new[cycle_index+1:cycle_index+1+future_horizon]

    L=min(len(true_soh_future),len(pred_soh))

    r2=r2_score(true_soh_future[:L],pred_soh[:L])

    mae=mean_absolute_error(true_soh_future[:L],pred_soh[:L])

    rmse=np.sqrt(np.mean((true_soh_future[:L]-pred_soh[:L])**2))

    print(os.path.basename(file_path),"-> R2:",r2," MAE:",mae," RMSE:",rmse)


print("\nTesting results\n")

for f in test_files:

    evaluate_on_file(f)
````
